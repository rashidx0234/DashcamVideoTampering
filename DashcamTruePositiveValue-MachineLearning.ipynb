```python
combine = A400s+A500s+A800s+D08+HuaweiNova3i+Iphone12ProMax+N3Pro+SamsungA23+Z40

arr1 = np.full(50, 0)
arr2 = np.full(50, 1)
arr3 = np.full(50, 2)
arr4 = np.full(50, 3)
arr5 = np.full(50, 4)
arr6 = np.full(50, 5)
arr7 = np.full(50, 6)
arr8 = np.full(50, 7)
arr9 = np.full(50, 8)

label = [*arr1, *arr2, *arr3, *arr4, *arr5, *arr6, *arr7, *arr8, *arr9]

len(combine)
```

```python
plt.imshow(combine[285]), plt.axis('off')
label[20]
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
```

```python
combine = np.array(combine)
label = np.array(label)

combine = (combine.reshape(combine.shape[0],combine.shape[1]*combine.shape[2]))

X_train,X_test,y_train,y_test = train_test_split(combine,label,test_size=0.3,random_state=42,shuffle=True)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
mlp = MLPClassifier(hidden_layer_sizes = (200,100), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
```

```python
predict = mlp.predict(X_test)
predict
```

```python
cm = metrics.confusion_matrix(y_test, predict)
score = mlp.score(X_test, y_test)

import seaborn as sns

plt.figure(figsize=(5,5))

sns.heatmap(cm, annot=True, fmt='d',
           linewidth=0.5, square=True, cmap='GnBu_r')

plt.xlabel("Predicted label",fontsize=10)
plt.ylabel("Actual label",fontsize=10)
plt.title("Accuracy Score: {0}".format(score),fontsize=10)
plt.show()
```

```python
metrics.accuracy_score(y_test,predict)
```

```python
print(metrics.classification_report(y_test,predict,labels=mlp.classes_.tolist()))
```
